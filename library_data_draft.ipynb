```python
def convert_time(time):
    if(time not in ["closed", "nan"]):
        out = datetime.strptime(time,"%I:%M%p")
        out = out.replace(year=datetime.now().year, day=datetime.now().day, month=datetime.now().month)
        return out
    else:
        return time



def process(x):
    x = str(x)
    x = x.split(" - ")
    x = [y.strip() for y in x]
    return {
        "opening":convert_time(x[0]),
        "closing":convert_time(x[-1])
    }

def clean_column(column):
    return [process(x) for x in list(column)]

clean_column(response["mn"])
```

```python
clean_data = pd.DataFrame(response["name"])
clean_data["lat"] = response["latitude"]
clean_data["lon"] = response["longitude"]
clean_data[0] = clean_column(response["mn"])
clean_data[1] = clean_column(response["tu"])
clean_data[2] = clean_column(response["we"])
clean_data[3] = clean_column(response["th"])
clean_data[4] = clean_column(response["fr"])
clean_data[5] = clean_column(response["sa"])
clean_data[6] = clean_column(response["su"])
clean_data
```

```python
def is_open(index, time=datetime.now()):
    open_time = clean_data.loc[index,time.weekday()]["opening"]
    close_time = clean_data.loc[index,time.weekday()]["closing"]
    if (open_time not in ["closed", "nan"]) and (close_time not in ["closed", "nan"]):
        has_opened = time > open_time
        has_not_closed = time < close_time
        return (has_opened and has_not_closed)
    else:
        return False
```
